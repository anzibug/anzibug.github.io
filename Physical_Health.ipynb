```python
# Pre-programmed user responses (modify as needed)
user_feelings = [
    "I've been feeling tired and sluggish lately.",
    "I'm trying to lose some weight and need a safe exercise plan.",
    "I'm recovering from an injury and need an injury-friendly workout."
]

user_activity_level = [
    "I don't exercise much right now.",
    "I go for a jog three times a week."
]

user_medical_conditions = [
    "I don't have any major medical conditions.",
    "I have asthma, but it's well-controlled with medication."
]

user_dietary_preferences = [
    "I mostly eat a balanced diet, but struggle with portion control.",
    "I'm a vegetarian."
]

user_goals = [
    "I'd like to feel more energized and lose 10 pounds.",
    "I want to build muscle and improve my fitness."
]


def greet_user():
  """Greets the user and introduces GrowWell."""
  print("Hello! I'm GrowWell, your AI assistant for overall well-being.")
  print("I'm here to help you reach your physical health goals. How are you feeling today?")

def get_user_feelings():
  """Prompts user to share their feelings and offers pre-programmed choices."""
  print("Here are some common concerns people have. How are you feeling?")
  for i, feeling in enumerate(user_feelings):
    print(f"{i+1}. {feeling}")
  choice = int(input("Enter a number corresponding to your feelings (or type your own): "))
  if 0 < choice <= len(user_feelings):
    return user_feelings[choice-1]
  else:
    return input("Tell me more about how you're feeling: ")

def get_initial_assessment():
  """Asks user for details and offers pre-programmed choices."""
  print("To personalize your plan, I'll need some information:")
  activity_level = input("How often do you exercise per week? (Choose a number or type your own): ")
  for i, level in enumerate(user_activity_level):
    print(f"{i+1}. {level}")
  print("Do you have any pre-existing medical conditions? (yes/no): ")
  medical_conditions = input(" (y/n): ")
  if medical_conditions.lower() == "y":
    medical_conditions = input("Tell me more about your medical conditions: ")
  else:
    medical_conditions = "None"
  print("Do you have any dietary preferences? (vegetarian, vegan, etc.): ")
  dietary_preferences = input(" (e.g., vegetarian, no preference): ")
  print("What are your specific goals? (weight loss, muscle building, etc.): ")
  for i, goal in enumerate(user_goals):
    print(f"{i+1}. {goal}")
  choice = int(input("Enter a number corresponding to your goal (or type your own): "))
  if 0 < choice <= len(user_goals):
    return activity_level, medical_conditions, dietary_preferences, user_goals[choice-1]
  else:
    return activity_level, medical_conditions, dietary_preferences, input("Tell me more about your goals: ")

def main():
  greet_user()
  feelings = get_user_feelings()
  print(f"Thanks for sharing that you're feeling {feelings}.")
  activity_level, medical_conditions, dietary_preferences, goals = get_initial_assessment()
  print(f"Thanks for providing this information. Now I have a better understanding of your current situation and goals (activity level: {activity_level}, medical conditions: {medical_conditions}, dietary preferences: {dietary_preferences}, goals: {goals}).")
  # Phase 2 will be implemented in a separate script.

if __name__ == "__main__":
  main()
```
