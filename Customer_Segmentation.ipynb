Elbow Method to Determine Optimal K Value

```python
X1=df.loc[:,["Age","Spending Score (1-100)"]].values

from sklearn.cluster import KMeans
WCSS = []
for k in range(1,11):
  kmeans = KMeans(n_clusters=k, init='k-means++')
  kmeans.fit(X1)
  WCSS.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),WCSS, linewidth=2,color="green", marker="8")
plt.xlabel("K value")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
lable = kmeans.fit_predict(X1)
print(lable)
```

```python
print(kmeans.cluster_centers_)
```

Customer Segmentation based on Age and Spending Score

```python
plt.scatter(X1[:,0], X1[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.title('Cluster of Customers')
plt.xlabel('Age')
plt.ylabel('Spending Score(1-100)')
plt.show()
```

```python
X1=df.loc[:,["Age","Spending Score (1-100)"]].values

from sklearn.cluster import KMeans
WCSS = []
for k in range(1,11):
  kmeans = KMeans(n_clusters=k, init='k-means++')
  kmeans.fit(X1)
  WCSS.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),WCSS, linewidth=2,color="green", marker="8")
plt.xlabel("K value")
plt.ylabel("WCSS")
plt.show()
```

Elbow Method to Determine Optimal Number of Clusters for Customer Segmentation based on Annual Income and Spending Score.

```python
X2=df.loc[:,["Annual Income (k$)","Spending Score (1-100)"]].values

from sklearn.cluster import KMeans
WCSS = []
for k in range(1,11):
  kmeans = KMeans(n_clusters=k, init='k-means++')
  kmeans.fit(X2)
  WCSS.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),WCSS, linewidth=2,color="green", marker="8")
plt.xlabel("K value")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5)
lable = kmeans.fit_predict(X2)
print(lable)
```

```python
print(kmeans.cluster_centers_)
```

Clusters of Customers based on Annual Income and Spending Score

```python
plt.scatter(X2[:,0], X2[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.title('Cluster of Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score(1-100)')
plt.show()
```

```python
X3=df.iloc[:,1:]

WCSS = []
for k in range(1,11):
  kmeans = KMeans(n_clusters=k, init='k-means++')
  kmeans.fit(X3)
  WCSS.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),WCSS, linewidth=2,color="green", marker="8")
plt.xlabel("K value")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5)
lable = kmeans.fit_predict(X3)
print(lable)
```

```python
print(kmeans.cluster_centers_)
```

3D Clustering of Customers based on Age, Annual Income, and Spending Score

```python
clusters = kmeans.fit_predict(X3)
df["lable"] = clusters

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df.Age[df.lable == 0], df["Annual Income (k$)"][df.lable == 0],df["Spending Score (1-100)"][df.lable == 0], c='red', s=60)
ax.scatter(df.Age[df.lable == 1], df["Annual Income (k$)"][df.lable == 1],df["Spending Score (1-100)"][df.lable == 1], c='blue', s=60)
ax.scatter(df.Age[df.lable == 2], df["Annual Income (k$)"][df.lable == 2],df["Spending Score (1-100)"][df.lable == 2], c='green', s=60)
ax.scatter(df.Age[df.lable == 3], df["Annual Income (k$)"][df.lable == 3],df["Spending Score (1-100)"][df.lable == 3], c='yellow', s=60)
ax.scatter(df.Age[df.lable == 4], df["Annual Income (k$)"][df.lable == 4],df["Spending Score (1-100)"][df.lable == 4], c='orange', s=60)
ax.set_zlabel('Spending Score (1-100)')

plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
ax.set_zlabel('Spending Score (1-100)')

plt.show()
```
