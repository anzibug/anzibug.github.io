```python
import datetime

class Habit:
    def __init__(self, name, reminder):
        self.name = name
        self.reminder = reminder
        self.completions = []

    def add_completion(self, completion):
        self.completions.append(completion)

    def get_completion_message(self):
        if all(completion == "yes" for completion in self.completions):
            return "Completed a task!"
        else:
            return "Need to work hard!"

class HabitTracker:
    def __init__(self):
        self.habits = []
        self.streak = 0

    def add_habit(self):
        name = input("Enter habit name: ")
        reminder = input("Enter habit reminder: ")
        habit = Habit(name, reminder)
        self.habits.append(habit)

    def mark_habit_completion(self):
        completion_list = []
        for habit in self.habits:
            completion = input(f"Did you complete the habit '{habit.name}' today? (yes/no): ")
            completion_list.append(completion.lower())
        for habit, completion in zip(self.habits, completion_list):
            habit.add_completion(completion)

    def display_habits(self):
        for index, habit in enumerate(self.habits):
            print(f"{index + 1}. {habit.name} - Reminder: {habit.reminder}")

    def calculate_streak(self):
        if all(habit.get_completion_message() == "Completed a task!" for habit in self.habits):
            self.streak += 1
        else:
            self.streak = 0

    def display_completion_message(self):
        for habit in self.habits:
            message = habit.get_completion_message()
            print(f"{habit.name}: {message}")

    def display_streak(self):
        print(f"Current streak: {self.streak} days")

# Example usage:
tracker = HabitTracker()

# Add habits
for _ in range(5):
    tracker.add_habit()

# Track habits for 4 weeks
for _ in range(4 * 7):
    # Display habits
    tracker.display_habits()

    # Mark habit completions
    tracker.mark_habit_completion()

    # Calculate streak
    tracker.calculate_streak()

    # Display completion messages
    tracker.display_completion_message()

    # Display streak
    tracker.display_streak()
    print()
yye
```
